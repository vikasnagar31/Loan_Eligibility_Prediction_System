### Handling null values

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
cat_col=['Gender','Dependents','Self_Employed']
for col in cat_col:
    mode=df[col].mode().values[0]
    df[col]=df[col].fillna(mode)
```

```python
num_col=['LoanAmount','Loan_Amount_Term','Credit_History']
for col in num_col:
    median=df[col].median()
    df[col]=df[col].fillna(median)
    
```

```python
df.isna().sum()
```

### Converting Categorical to numeric

```python
mapping_df=pd.read_csv('mapping_dict_train.csv')
mapping_df
```

```python
mapping_dict=dict(mapping_df.values)
mapping_dict
```

```python
df.dtypes
```

```python
cat_col = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']

for col in cat_col:
   df[col] = df[col].apply(lambda x: mapping_dict[x])
 
 
```

```python
df
```

```python
final_df=pd.DataFrame(df['Loan_ID'])
final_df.head()
```

```python
df=df.iloc[:,1:]
df.head()
```
